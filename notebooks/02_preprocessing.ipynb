```python
import re
import pandas as pd
import unicodedata
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download("stopwords")
stops = set(stopwords.words("english"))
```

```python
df=pd.read_csv('../Data/raw/data.csv')
```

```python
df.head()
```

```python
def normalize_text(text):
    text=unicodedata.normalize('NFKD',text)
    text=''.join(c for c in text if not unicodedata.combining(c))

    text=text.lower()
    return text
```

```python
def tokenize(text):
    return word_tokenize(text)
```

```python
punct_regex = re.compile(r"^[a-zA-Z0-9]+$")  # garder uniquement tokens alphanumériques

def remove_punctuation_tokens(tokens):
    return [t for t in tokens if punct_regex.match(t)]
```

```python
def remove_stopwords_tokens(tokens):
    return [t for t in tokens if t not in stops]
```

```python
def remove_duplicates_tokens(tokens):
    seen = set()
    unique = []
    for t in tokens:
        if t not in seen:
            unique.append(t)
            seen.add(t)
    return unique
```

```python
def preprocess_nltk(text):
    # 1) Normalisation
    text = normalize_text(text)
    
    # 2) Tokenisation
    tokens = tokenize(text)
    
    # 3) Remove punctuation
    tokens = remove_punctuation_tokens(tokens)
    
    # 4) Remove stopwords
    tokens = remove_stopwords_tokens(tokens)
    
    # 5) Remove duplicates
    tokens = remove_duplicates_tokens(tokens)
    
    return tokens
```

```python
df["tokens"] = df["text"].apply(preprocess_nltk)
df[["text", "tokens"]].head()
```
